```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('customer_analytics.csv')
print("Head of the dataset : \n",df.head(), "\n\n")
print("\n\nInformation of Dataset : \n",df.info(), "\n\n")
print("Description of the dataset : \n",df.describe(), "\n\n")
```

```python
df.isnull().sum()
df.duplicated().sum()
df.drop_duplicates()
```

```python
df['Education'] = df['Education'].fillna(df['Education'].mode()[0])
df['AnnualIncome'] = df['AnnualIncome'].fillna(df['AnnualIncome'].mean())
print("-> Filling the null values with mean,median,mode values in the 'Education' and 'AnnualIncome', which nearly have 5% of null values")
df.isnull().sum()
```

```python
# Univariate Plots...
print('UNIVARIATE VARIABLES')
sns.histplot(df['Age'],kde=True)
plt.title('Histplot for AGE')
plt.show()
print("-> Here, histplot used to show the distribution of the data(of AGE column), which heps us to understand the frequency and spread of the data.\n")
plt.boxplot(df['AnnualIncome'])
plt.title('Boxplot for Annual Income')
plt.show()
print("-> Here, Boxplot show the spread of a numerical variable(for AnnualIncome Column) and helps identify median, range, and outliers.\n ")
sns.countplot(df['Education'])
plt.title('Count Plot for Education')
plt.show()
print("-> Countplot show the frequency of each category in categorical variables.\n")
```

```python
#Bivariate variable
print('BIVARIATE VARIABLES')
sns.scatterplot(x = "AnnualIncome", y = "SpendingScore", data = df)
plt.title('Scatter Plot for AnnualIncome and Spendingscore')
plt.show()
print("->  Scatterplot shows how two numeric variables(here, for AnnualIncome and SpendingScore columns) are related.\n")

sns.boxplot(x = "Education", y = "AnnualIncome", data = df)
plt.title('Boxplot Education and AnnualIncome')
plt.show()
print("->  Boxplot shows how a numeric variable(here, Education and AnnualIncome Columns) changes across categories and helps identify median, range, and outliers ")
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True, cmap='coolwarm')
plt.show()
print("->  Heatmap shows relationships between numeric variables using colors.")
```
