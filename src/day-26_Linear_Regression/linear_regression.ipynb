```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
X = df[['year']]
y = df['per capita income (US$)']
```

```python
plt.xlabel('Year')
plt.ylabel('per capita income (US$)')
plt.plot(X,y,color = 'red', marker='*')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
new_data = pd.DataFrame({'year' : [2017,2018,2019,2020]})
prediction = reg.predict(new_data)
prediction = np.round(prediction,2)
prediction
```

```python
new_data['per capita income (US$)'] = prediction
```

```python
new_data.to_csv('canada_per_capita_income.csv',mode = "a",index = False,header=False)
```
